```python
import pandas as pd

imgs = pd.read_csv('data/img.csv')

imgs.head()
```

```python
imgs.shape
```

```python
import matplotlib.pyplot as plt

im = imgs.iloc[:-1, 1:]

plt.imshow(im, aspect='auto', cmap='inferno')
plt.show()
```

# Resized image

```python
import cv2

height = int(im.shape[0])
new_size = (150, height)

resized_image = cv2.resize(im.to_numpy(), new_size, interpolation=cv2.INTER_AREA)

plt.imshow(resized_image, aspect='auto', cmap='inferno')
```
